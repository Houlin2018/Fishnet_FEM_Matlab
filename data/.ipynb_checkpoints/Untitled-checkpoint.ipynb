```python
# Let's create a function to process multiple files and plot them together.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def plot_weibull_scale(file_list, labels):
    plt.figure(figsize=(10, 6))

    for file, label in zip(file_list, labels):
        # Load the data
        data = pd.read_csv(file)
        
        # Assuming the strength data is in the first column
        strength = data.iloc[:, 0].values
        
        # Sort the strength values
        strength_sorted = np.sort(strength)
        
        # Weibull probability calculation
        weibull_prob = np.linspace(1/(len(strength_sorted)+1), 1, len(strength_sorted))
        
        # Calculate the Weibull scale (ln(-ln(1-pdf)))
        weibull_scale = np.log(-np.log(1 - weibull_prob))
        
        # Plotting
        plt.plot(strength_sorted, weibull_scale, label=label)
    
    plt.xlabel('Strength', fontsize=16)
    plt.ylabel(r'$\ln(-\ln(1-\text{pdf}))$', fontsize=16)
    plt.title('Weibull Scale vs. Strength')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage with the two files you provided
file_list = [
    './sqnom_strength_10_10_WG_0p1.csv',
    './sqnom_strength_10_10_WG_0p05.csv'
]
labels = ['0p1', '0p05']

plot_weibull_scale(file_list, labels)
```
